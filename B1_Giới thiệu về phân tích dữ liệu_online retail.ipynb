```python
#Question2:How many orders have been sold and what is the total revenue?
def cal_revenue(df):
    price, quantinty = df['UnitPrice'],df['Quantity']
    return np.sum(price*quantinty)
```

```python
total_revenue = cal_revenue(df)
total_revenue
```

```python
#Question3:Top 10 best seller products
top_10_product = df.groupby(by =['StockCode','Description']).agg({'Quantity':sum}).reset_index()
top_10_product = top_10_product.sort_values('Quantity',ascending = False)
top_10_product.head(10)
```
